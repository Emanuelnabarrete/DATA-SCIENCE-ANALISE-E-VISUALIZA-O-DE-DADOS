```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
linguas = movies.original_language.unique()
```

```python
linguas
```

```python
qnt_linguas = movies.original_language.value_counts()
```

```python
qnt_linguas
```

```python
df_linguas = qnt_linguas.to_frame().reset_index()
df_linguas.columns = ["Linguas" , "Total"]
df_linguas.head()
```

Grafico barras do seaborn || dizer o x , y , e dados(data) para cria-lo

```python
sns.barplot(x = "Linguas" , y = "Total", data = df_linguas)#*****************************
```

Grafico de barras mais comum - > trazer o x, o tipo que ira realiza (kind) , os dados (data) || forma mais facil de usar

```python
sns.catplot(x="original_language", kind="count", data=movies)
```

Grafico de torta do **matplotlib**

```python
plt.pie(df_linguas["Total"], labels =df_linguas['Linguas'])
```

```python
total_lingua = movies["original_language"].value_counts()
```

```python
total_lingua
```

**.loc[" "]** localiza e traz o total de todos os valores que estao dentro

```python
en = total_lingua.loc["en"]
```

```python
en
```

```python
rest = (total_lingua.sum()) - en
```

```python
print(en , rest)
```

Criando dicionario - > Criando df a partir dele

```python
dados= {
        "Linguas" : ["ingles" , "outros"],
        "total" : [en , rest]
        }

dados = pd.DataFrame(dados)
dados
```

plotando dicionario

```python
sns.barplot(x = "Linguas" , y = "total" , data = dados)
```

not_en = filmes - > linguas(original_language) -> (linguas diferente de ingles) - > so as linguas - > liguas categorizadas por quantidade de vezes que apareceu

total_sem_ingles = total de filmes sem o ingles ( trago o df )

```python
not_en = movies.query("original_language != 'en'").original_language.value_counts()
total_sem_en = movies.query("original_language != 'en'")
```

sns.catplot(x = original_language( linguas ) , kind  = "count" ( quantas vezes apareceu) , data = dados (total_sem_ingles)







aspect = tamanho (semelhante a largura) | order (pondo em ordem) = **not_en** com ele podemos la em cima tirar a categorica e trazer para ca (plot) | palette = Cores

```python
sns.catplot(x="original_language", kind="count", data= total_sem_en, aspect = 2, order =not_en.index , palette = "PuOr")
```

Importando e usando numpy(arrays)
